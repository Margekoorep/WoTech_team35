```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pokemon_df = pd.read_csv('/content/Pokemon.csv')
grass_df=pokemon_df[pokemon_df['Type 1']=='Grass']
water_df=pokemon_df[pokemon_df['Type 1']=='Water']
```

```python
sns.regplot(x='Attack', y='Defense', data=grass_df, color='green', ci=99, marker='x', line_kws={'color':'red'})
plt.title('Correlation of Attack and Defense')
correlation1 = grass_df['Attack'].corr(grass_df['Defense'])
print(f'The pearson correlation for grass type pokemon between attack and defence is {correlation1}')
```

```python
sns.regplot(x='Attack', y='Defense', data=water_df, color='blue', ci=99, marker='x', line_kws={'color':'red'})
plt.title('Correlation of Attack and Defense')
correlation2 = water_df['Attack'].corr(water_df['Defense'])
print(f'The pearson correlation for water type pokemon between attack and defence is {correlation2}')
```
